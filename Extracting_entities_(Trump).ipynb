**Merging the Mentioning files for comparison**

```python
data1=pd.read_csv("/content/@mentions_list(trump_NP).csv")
data1.head()
```

```python
data2=pd.read_csv("/content/@mentions_list(trump_P).csv")
data2.head()
```

```python
# Rename the column
data1.rename(columns={'frequency': 'freq.before presidency'},inplace=True)
```

```python
# Rename the column
data2.rename(columns={'frequency': 'freq.during pesidency'},inplace=True)
```
